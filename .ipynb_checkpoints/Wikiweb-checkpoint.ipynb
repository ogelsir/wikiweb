```python
import urllib.request
from bs4 import BeautifulSoup as bs
import re
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Page:
    '''
    represents a single wikipedia page
    '''
    def __init__(self, myName, myTier, myStop):
        '''
        constructor for page class
        @param myName = name of article
        @param myTier = current tier/level
        @param myStop = last tier to include
        '''
        self.name = myName # name of article
        self.tier = myTier # current tier/level
        self.stop = myStop # tier to not include
        
        self.nameList = [] # contains strings representing names of articles
        self.pgDict = {} # key = strings, value = Pages
        self.scoreList = {} # key = strings, value = floats
        
        if( self.tier + 1 < self.stop ):
            response = urllib.request.urlopen('https://en.wikipedia.org/wiki/' + self.name)
            html = response.read()
            soup = bs(html,'lxml')
            for par in soup.find_all('p'): # finds all suitable articles
                for link in par.find_all('a', href=True) :
                    url = link.get('href')
                    if( (url.startswith('/wiki/')) and (':' not in url) ):
                        if( '#' in url ):
                            hashInd = url.find('#')
                            curSubj = url[6:hashInd]
                        else:
                            curSubj = url[6:]
                        curPage = Page(curSubj, self.tier + 1, self.stop)
                        self.nameList.append(curSubj)
                        self.pgDict[curSubj] = curPage
        for s in self.nameList: # fills scores
            self.scoreList[s] = self.getScore(s)
    
    def getScore(self, aPage):
        '''
        returns a float between 0 and 1 representing the correlation between self and aPage
        @param aPage is another Page object

        score is calculated by len(intersection)/ avg(len(self),len(aPage))
        '''
        selfList = self.getNameList()
        aList = self.getPgDict()[aPage].getNameList()
        intersect = len([x for x in selfList if x in aList])
        selfLen = len(selfList)
        otherLen = len(aList)
        score = intersect/((selfLen + otherLen)/2)
        return score  
    
    # getter variables
    def getName(self):
        return self.name
    def getTier(self):
        return self.tier
    def getStop(self):
        return self.stop
    def getNameList(self):
        return self.nameList
    def getPgDict(self):
        return self.pgDict
    def getScoreList(self):
        return self.scoreList
    # tostring
    def __str__(self):
        return self.name
```
